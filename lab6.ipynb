```python
import boto3
from botocore.exceptions import ClientError
import json

def get_secret(secret_name):
    region_name = "us-east-1"

    # Create a Secrets Manager client
    session = boto3.session.Session()
    client = session.client(
        service_name='secretsmanager',
        region_name=region_name
    )

    try:
        get_secret_value_response = client.get_secret_value(
            SecretId=secret_name
        )
    except ClientError as e:
        raise e

    secret = get_secret_value_response['SecretString']
    
    return json.loads(secret)
```

```python
import psycopg2

conn = psycopg2.connect(host=get_secret('postgresql')['host'],
                      user=get_secret('postgresql')['username'],
                      password = get_secret('postgresql')['password'],
                      dbname=get_secret('postgresql')['engine'])

cur = conn.cursor()
```

```python
sql = """
     select * from income inner join name on name.fips=income.fips
     """
```

```python
import pandas as pd

df = pd.read_sql(sql, con=conn)
df.head() #show the top 5 rows
```

```python
import matplotlib.pyplot as plt

# Sort values by name and year, reset index
df = df.sort_values(['name', 'year']).reset_index(drop=True)

# Calculate year-over-year income growth rate for each state
df['income_growth_rate'] = df.groupby('name')['income'].pct_change()

# Filter out the data for the last year
df_last_year = df[df['year'] == df['year'].max()]

# Create a bar chart to show the income growth rate for the last year
df_last_year.plot(x='name', y='income_growth_rate', kind='bar', figsize=(10,6))
plt.title('Year-Over-Year Income Growth Rate by State')
plt.xlabel('State')
plt.ylabel('Income Growth Rate')
plt.show()
```

```python
# Find the row with the highest income
highest_income = df[df['income'] == df['income'].max()]

# Print the state and year with the highest income
print(highest_income[['name', 'year']])
```

```python
# Calculate the average year-over-year income growth rate for each state
average_growth_rate = df.groupby('name')['income_growth_rate'].mean().reset_index()

# Create a bar chart to show the average income growth rate by state
average_growth_rate.plot(x='name', y='income_growth_rate', kind='bar', figsize=(10,6))
plt.title('Average Year-Over-Year Income Growth Rate by State')
plt.xlabel('State')
plt.ylabel('Average Income Growth Rate')
plt.show()
```

```python
%%ai gpt4 -f code
Calcuate the years with the highest income and visualize it in a pie chart
```

```python
import matplotlib.pyplot as plt

# Calculate sum of income for each year
income_per_year = df.groupby('year')['income'].sum()

# Find the year with the highest total income
highest_income_year = income_per_year.idxmax()

# Create a pie chart with the year with the highest income highlighted
explode = [0.1 if year == highest_income_year else 0 for year in income_per_year.index]
plt.pie(income_per_year, explode=explode, labels=income_per_year.index, autopct='%1.1f%%')
plt.title('Income Distribution by Year')

# Display the plot
plt.show()
```
